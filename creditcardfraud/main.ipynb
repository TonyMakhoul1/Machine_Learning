```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

minmaxscaler = MinMaxScaler()
standardscaler = StandardScaler()

dataset['Amount'] = standardscaler.fit_transform(dataset[['Amount']])
dataset['Time'] = minmaxscaler.fit_transform(dataset[['Time']])
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
label = dataset['Class']
```

```python
dataset.drop(columns='Class',inplace=True)
```

```python

from sklearn.decomposition import PCA

pca = PCA(n_components=20)
pca_data = pca.fit_transform(dataset)
pca.explained_variance_
```

```python

plt.step(list(range(1, len(np.cumsum(pca.explained_variance_ratio_)) + 1)), np.cumsum(pca.explained_variance_ratio_))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Eigen Values')
plt.ylabel('Ratio of Variance Explained')
plt.title('Variance Covered by each Eigen Value')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.001,
    max_samples=256,
    random_state= 42
)

iso_forest.fit(pca_data)

dataset['anomaly'] = iso_forest.predict(pca_data)
```

```python
dataset['anomaly'].value_counts()
```

```python
normal = dataset[dataset['anomaly'] == 1]
anomalies = dataset[dataset['anomaly'] == -1]
```

```python
plt.figure(figsize=(10,6))

plt.scatter(normal['Time'], normal['Amount'], c='blue')
plt.scatter(anomalies['Time'], anomalies['Amount'], c='orange')
plt.xlabel('Time')
plt.ylabel('Transaction Amount')
plt.show()
```

```python
dataset['predicted'] = dataset['anomaly'].map({1:0, -1:1})
```

```python
dataset.head()
```

```python
dataset['label'] = label
```

```python
dataset.head()
```

```python
from sklearn.metrics import precision_score, f1_score, recall_score

precision = precision_score(label, dataset['predicted'])
recall = recall_score(label, dataset['predicted'])
f1 = f1_score(label, dataset['predicted'])

print(precision)
print(recall)
print(f1)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(label, dataset['predicted'])
print(acc)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(label, dataset['predicted'])
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
# The precision, recall, f1 score are low due to the imbalanced data
# Many of fraud transactions are predicted as normal transaction
# The accuracy score is very high 
# the flagged transactions are the fraud transactions (identified as outliers)
```
