```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
car_dataset = pd.read_csv("./data/car data.csv")
```

```python
car_dataset.Seller_Type.value_counts()
```

```python
car_dataset.Transmission.value_counts()
```

```python
# Label Encoding
# Fuel_type
car_dataset.replace({"Fuel_Type": {"Petrol":0, "Diesel":1, "CNG":2}}, inplace=True)

# Seller_Type
car_dataset.replace({"Seller_Type": {"Dealer":0, "Individual":1}}, inplace=True)

# Transmission
car_dataset.replace({"Transmission": {"Manual":0, "Automatic":1}}, inplace=True)
```

```python
car_dataset.head()
```

```python
X = car_dataset.drop(columns=["Car_Name","Selling_Price"], axis=1)
y = car_dataset["Selling_Price"] 
```

```python
X
```

```python
y
```

Training and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=2)
```

Linear Regression Model

```python
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train, y_train)
```

Model Evaluation

```python
training_data_pred = lin_reg_model.predict(X_train)
```

```python
# R squared Error
error_score = metrics.r2_score(y_train, training_data_pred)
# R squared should be low 
error_score
```

```python
plt.scatter(y_train, training_data_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices Vs Predicted Prices")
plt.show()
```

```python
test_data_pred = lin_reg_model.predict(X_test)
```

```python
# R squared error
error_score_test = metrics.r2_score(y_test, test_data_pred)
error_score_test
```

```python
plt.scatter(y_test, test_data_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices Vs Predicted Prices")
plt.show()
```

Lasso Regression

```python
lass_reg_model = Lasso()
```

```python
lass_reg_model.fit(X_train, y_train)
```

```python
training_data_pred = lass_reg_model.predict(X_train)
```

```python
error_score = metrics.r2_score(y_train, training_data_pred)
error_score
```

```python
plt.scatter(y_train, training_data_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices Vs Predicted Prices")
plt.show()
```

```python
test_data_pred = lass_reg_model.predict(X_test)
```

```python
error_score_test = metrics.r2_score(y_test, test_data_pred)
error_score_test
```

```python
plt.scatter(y_test, test_data_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices Vs Predicted Prices")
plt.show()
```
